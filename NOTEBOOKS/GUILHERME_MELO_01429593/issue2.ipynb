```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# 1. Classe
print("----- Classe -----")
print(vinhos['classe'].value_counts())

# 2. Alcool, Acido Malico, Magnesio e Matiz
print("\n----- Descrição Estatística -----")
print(vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe())
# --- 2. Criação de Boxplots ---

# Definir estilo dos gráficos Seaborn
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))  # Ajustar tamanho da figura

# Boxplot para 'magnesio'
plt.subplot(2, 2, 1)  # Criar uma grade 2x2 de subplots
sns.boxplot(y='magnesio', data=vinhos)
plt.title('Boxplot do Magnésio')

# Boxplot para 'prolina'
plt.subplot(2, 2, 2)
sns.boxplot(y='prolina', data=vinhos)
plt.title('Boxplot da Prolina')

# Boxplot para 'cinzas'
plt.subplot(2, 2, 3)
sns.boxplot(y='cinzas', data=vinhos)
plt.title('Boxplot das Cinzas')

# Boxplot para 'alcool'
plt.subplot(2, 2, 4)
sns.boxplot(y='alcool', data=vinhos)
plt.title('Boxplot do Álcool')

# Ajustar o layout para evitar sobreposição
plt.tight_layout()
plt.show()

# --- 3. Boxplots Segregados por Classe ---

plt.figure(figsize=(12, 8))  # Ajustar o tamanho da figura

plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe', ax=plt.gca())
plt.title('Magnésio por Classe')

plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe', ax=plt.gca())
plt.title('Fenois Totais por Classe')

plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe', ax=plt.gca())
plt.title('Ácido Málico por Classe')

plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe', ax=plt.gca())
plt.title('Álcool por Classe')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})


# Histograma para todos os dados da coluna 'fenois_totais'
plt.figure(figsize=(8, 5)) 
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True)
plt.title('Histograma - Fenois Totais (Todas as Classes)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma para 'fenois_totais' apenas na classe 3
plt.figure(figsize=(8, 5))
sns.histplot(vinhos[vinhos['classe'] == 3]['fenois_totais'], bins=10, kde=True)
plt.title('Histograma - Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
