Faça a leitura dos dados sobre vinhos da UCI, utilizando o seguinte código:

```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

2. Criação de Boxplots
Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:
*   magnesio
*   prolina
*   cinzas
*   alcool

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
sns.boxplot(x=vinhos['magnesio'])
plt.title('Boxplot de Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(x=vinhos['prolina'])
plt.title('Boxplot de Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(x=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(x=vinhos['alcool'])
plt.title('Boxplot de Álcool')

plt.tight_layout()
plt.show()
```

3. Boxplots Segregados por Classe
Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:

*   magnesio
*   fenois_totais
*   acido_malico
*   alcool


```python
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de Magnésio por Classe')
plt.suptitle('')  # Remove o título padrão do subplot
plt.show()

vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenóis Totais por Classe')
plt.suptitle('')  # Remove o título padrão do subplot
plt.show()

vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ácido Málico por Classe')
plt.suptitle('')  # Remove o título padrão do subplot
plt.show()

vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')  # Remove o título padrão do subplot
plt.show()
```

4. Criação de Histograma
Crie um histograma para visualizar a distribuição dos dados na coluna:
fenois_totais, utilizando 10 bins

Dica: Utilize plt.hist() ou sns.histplot() para criar histogramas e ajuste o número de bins usando o argumento bins=10.
Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.
Dica: Filtre os dados com vinhos[vinhos['classe'] == '3'] antes de gerar o histograma.

```python
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.title('Histograma de Fenóis Totais')
plt.show()

# --------------- Apenas a classe 3 --------------------
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10)
plt.xlabel('Fenóis Totais (Classe 3)')
plt.ylabel('Frequência')
plt.title('Histograma de Fenóis Totais (Classe 3)')
plt.show()
```
