```python
import pandas as pd
import matplotlib.pyplot as plt
dados = pd.read_csv('https://archive.ics.uci.edu/static/public/53/data.csv')
dados
```

```python
#contando os valores unicos:
dados['class'].value_counts()
```

```python
# estatÃ­stica descritiva dos valores
dados.describe()
```

```python
# prompt: faÃ§a o df.describe segragado por cada valor unico de classe

for classe in dados['class'].unique():
  print(f"DescriÃ§Ã£o para a classe: {classe}")
  print(dados[dados['class'] == classe].describe())
  print("\n")
```

```python
# prompt: faÃ§a o histograma para sepal length de cada classe em class , apresente plots separada
for classe in dados['class'].unique():
  plt.figure()  # Cria uma nova figura para cada classe
  plt.hist(dados[dados['class'] == classe]['sepal length'], bins=10)
  plt.title(f'Histograma de Sepal Length para a classe: {classe}')
  plt.xlabel('Sepal Length')
  plt.ylabel('FrequÃªncia')
  plt.show()
```

```python
# prompt: gere o boxplot comparativo segregado por class da sepal length

import matplotlib.pyplot as plt

# Gere o boxplot comparativo segregado por class da sepal length
dados.boxplot(column=['sepal length'], by='class')
plt.title('Boxplot de Sepal Length por Classe')
plt.suptitle('') # Remove o tÃ­tulo padrÃ£o gerado pelo boxplot
plt.show()
```
