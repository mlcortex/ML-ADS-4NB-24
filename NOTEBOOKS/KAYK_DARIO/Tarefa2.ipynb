# **Exercício Prático: Análise Exploratória de Dados - Vinhos (UCI)**

### **Importando bibliotecas e dados**

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

# **3. Boxplots Segregados por Classe**

### **Magnesio**

```python
# boxplot da coluna magnesio, segregando os dados pela variável classe

plt.figure(figsize=(10, 6))
sns.boxplot(x='classe', y='magnesio', data=vinhos)
plt.title('Boxplot de Magnésio por Classe')
plt.xlabel('Classe')
plt.ylabel('Magnésio')
plt.show()
```

### **Fenois Totais**

```python
# boxplot da coluna magnesio, segregando os dados pela variável classe

plt.figure(figsize=(10, 6))
sns.boxplot(x='classe', y='fenois_totais', data=vinhos)
plt.title('Boxplot de Fenois Totais por Classe')
plt.xlabel('Classe')
plt.ylabel('Fenois Totais')
plt.show()
```

### **Acido Malico**

```python
# boxplot da coluna Acido Malico, segregando os dados pela variável classe

plt.figure(figsize=(10, 6))
sns.boxplot(x='classe', y='acido_malico', data=vinhos)
plt.title('Boxplot de Acido Malico por Classe')
plt.xlabel('Classe')
plt.ylabel('Acido Malico')
plt.show()
```

# **4. Criação de Histograma**

```python
# histograma para visualizar a distribuição dos dados na coluna: fenois_totais, utilizando 10 bins

plt.figure(figsize=(10, 6))
plt.hist(vinhos['fenois_totais'], bins=10)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```

### **CLASSE 3**

```python
# histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.

vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

plt.figure(figsize=(10, 6))
plt.hist(vinhos_classe_3['fenois_totais'], bins=10)
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
