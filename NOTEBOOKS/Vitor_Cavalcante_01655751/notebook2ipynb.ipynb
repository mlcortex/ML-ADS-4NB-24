```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
vinhos.head()
```

```python
# Contagem de valores por classe
vinhos['classe'].value_counts()
```

```python
vinhos[['alcool','acido_malico','magnesio','matiz']].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')

# criar boxplot
fig, axes = plt.subplots(2, 2, figsize=(15,10))

sns.boxplot(ax=axes[0, 0], data=vinhos, y='magnesio').set_title('Boxplot de Magnésio')
sns.boxplot(ax=axes[0, 1], data=vinhos, y='prolina').set_title('Boxplot de Prolina')
sns.boxplot(ax=axes[1, 0], data=vinhos, y='cinzas').set_title('Boxplot de Cinzas')
sns.boxplot(ax=axes[1, 1], data=vinhos, y='alcool').set_title('Boxplot de Álcool')

plt.tight_layout()
plt.show()
```

```python
# Boxplots segregados por classe
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.boxplot(ax=axes[0, 0], data=vinhos, x='classe', y='magnesio').set_title('Magnésio por Classe')
sns.boxplot(ax=axes[0, 1], data=vinhos, x='classe', y='fenois_totais').set_title('Fenois Totais por Classe')
sns.boxplot(ax=axes[1, 0], data=vinhos, x='classe', y='acido_malico').set_title('Ácido Málico por Classe')
sns.boxplot(ax=axes[1, 1], data=vinhos, x='classe', y='alcool').set_title('Álcool por Classe')

plt.tight_layout()
plt.show()
```

HISTOGRAMAS

```python
# Histograma de fenois_totais
plt.figure(figsize=(8,6))
sns.histplot(vinhos['fenois_totais'], bins=10)
plt.title('Distribuição de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```

```python
# Filtrar dados da classe 3
classe_3 = vinhos[vinhos['classe'] == '3']

# Histograma para fenois_totais na classe 3
plt.figure(figsize=(8,6))
sns.histplot(classe_3['fenois_totais'], bins=10)
plt.title('Distribuição de Fenois Totais - Classe 3')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
