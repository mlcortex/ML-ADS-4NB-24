```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

# 1. Análise Exploratória de Variáveis

```python
#classe: Faça a contagem das diferentes classes presentes no dataset.
print("Contagem das diferentes classes:")
print(vinhos['classe'].value_counts())
```

```python
# Alcool, Acido_melico, Magnesio, Matriz
print("\nEstatísticas descritivas para as colunas:")
print(vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe())
```

#2. Criação de Boxplots
Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:
magnesio,
prolina,
cinzas,
alcool.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x=vinhos['magnesio'])
plt.title('Boxplot - Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(x=vinhos['prolina'])
plt.title('Boxplot - Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(x=vinhos['cinzas'])
plt.title('Boxplot - Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(x=vinhos['alcool'])
plt.title('Boxplot - Álcool')

plt.tight_layout()
plt.show()
```

#3. Boxplots Segregados por Classe (com err, corrigir)

Crie um boxplot das seguintes colunas, segregando os dados pela classe variável:
magnesio,
fenois_totais,
acido_malico,
alcool.

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot - Magnésio por Classe')
plt.suptitle('Magnesio')

```

```python
plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot - Ácido Málcio por Classe')
plt.suptitle('Acido_Malico')
```

```python
plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot - Álcool por Classe')
plt.suptitle('Alcool')
```

#4. Criação de Histograma
Crie um histograma para visualizar a distribuição dos dados na coluna:

fenois_totais, utilizando 10 bins




```python
plt.figure(figsize=(15, 5))

# Histograma para fenois_totais
plt.subplot(1, 2, 1)
plt.hist(vinhos['fenois_totais'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma - Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
```

#Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins

```python
plt.subplot(1, 2, 2)
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histograma - Fenóis Totais (Classe 3)')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
