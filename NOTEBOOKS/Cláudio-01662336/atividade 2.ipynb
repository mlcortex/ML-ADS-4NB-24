```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar o dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
column_names_pt_br = [
    'classe', 'alcool', 'acido_malico', 'cinzas', 'alcalinidade_de_cinzas',
    'magnesio', 'fenois_totais', 'flavanoides', 'fenois_nao_flavanoides',
    'proantocianinas', 'intensidade_de_cor', 'matiz',
    'od280_od315_de_vinhos_diluidos', 'prolina'
]
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# Tarefa 1: Análise exploratória
print(vinhos['classe'].value_counts())
print(vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe())

# Tarefa 2: Boxplots
cols_boxplot = ['magnesio', 'prolina', 'cinzas', 'alcool']
for col in cols_boxplot:
    sns.boxplot(data=vinhos[col])
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.show()

# Tarefa 3: Boxplots segregados por classe
cols_by_class = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']
for col in cols_by_class:
    vinhos.boxplot(column=col, by='classe', grid=False)
    plt.title(f'Boxplot de {col} por Classe')
    plt.suptitle('')  # Remove o título padrão
    plt.xlabel('Classe')
    plt.ylabel(col)
    plt.show()

# Tarefa 4: Histogramas
sns.histplot(vinhos['fenois_totais'], bins=10, kde=False)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma apenas para a classe 3
classe_3 = vinhos[vinhos['classe'] == '3']
sns.histplot(classe_3['fenois_totais'], bins=10, kde=False)
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
