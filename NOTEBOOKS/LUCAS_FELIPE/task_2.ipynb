```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em portuguÃªs BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

```python
#boxplots da dispersÃ£o dos dados das colunas
#magnesio
#prolina
#cinzas
#alcool

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(data=vinhos, y='magnesio')
plt.title('DispersÃ£o de MagnÃ©sio')

plt.subplot(2, 2, 2)
sns.boxplot(data=vinhos, y='prolina')
plt.title('DispersÃ£o de Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(data=vinhos, y='cinzas')
plt.title('DispersÃ£o de Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(data=vinhos, y='alcool')
plt.title('DispersÃ£o de Ãlcool')

plt.tight_layout()

plt.show()
```

```python
#boxplots das colunas
#magnesio
#fenois_totais
#acido_malico
#alcool

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de MagnÃ©sio por Classe')
plt.suptitle('')

plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenois Totais por Classe')
plt.suptitle('')

plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ãcido MÃ¡lico por Classe')
plt.suptitle('')

plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Ãlcool por Classe')
plt.suptitle('')

plt.tight_layout()

plt.show()
```

```python
#dados da coluna fenois totais
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('FrequÃªncia')
plt.title('Histograma de Fenois Totais')
plt.show()
```

```python
# Filtrar os dados para a classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

plt.hist(vinhos_classe_3['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('FrequÃªncia')
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.show()
```
