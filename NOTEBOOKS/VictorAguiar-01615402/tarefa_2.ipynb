1. Análise Exploratória de Variáveis


```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# Análise da coluna 'classe'
classe_counts = vinhos['classe'].value_counts()
print("Contagem das Classes:")
print(classe_counts)

# Estatísticas descritivas para as colunas 'alcool', 'acido_malico', 'magnesio' e 'matiz'
estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print("\nEstatísticas Descritivas:")
print(estatisticas)
```

2. Criação de Boxplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Criar boxplots para as colunas
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(y='magnesio', data=vinhos)
plt.title('Boxplot de Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(y='prolina', data=vinhos)
plt.title('Boxplot de Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(y='cinzas', data=vinhos)
plt.title('Boxplot de Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(y='alcool', data=vinhos)
plt.title('Boxplot de Álcool')

plt.tight_layout()
plt.show()
```

3.Boxplots Segregados por Classe

```python
plt.figure(figsize=(15, 10))

# Criar boxplots segregados por classe
ax1 = plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe', ax=ax1)
ax1.set_title('Magnésio por Classe')

ax2 = plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe', ax=ax2)
ax2.set_title('Fenóis Totais por Classe')

ax3 = plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe', ax=ax3)
ax3.set_title('Ácido Málico por Classe')

ax4 = plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe', ax=ax4)
ax4.set_title('Álcool por Classe')

# Remover o título global que é adicionado automaticamente
plt.suptitle('')  # Remove o suptitle padrão
plt.tight_layout()
plt.show()
```

4. Criação de Histograma


```python
# Histograma de fenois_totais
plt.figure(figsize=(10, 6))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True)
plt.title('Distribuição de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma de fenois_totais para a classe 3
plt.figure(figsize=(10, 6))
sns.histplot(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10, kde=True)
plt.title('Distribuição de Fenóis Totais (Classe 3)')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
