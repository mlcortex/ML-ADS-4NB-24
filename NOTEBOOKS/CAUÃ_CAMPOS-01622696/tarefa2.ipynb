```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

```python
print("Estatísticas descritivas:")
print(vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe())

print("\nContagem das classes:")
print(vinhos['classe'].value_counts())
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.boxplot(vinhos['magnesio'])
plt.title('Magnésio')

plt.subplot(1, 4, 2)
plt.boxplot(vinhos['prolina'])
plt.title('Prolina')

plt.subplot(1, 4, 3)
plt.boxplot(vinhos['cinzas'])
plt.title('Cinzas')

plt.subplot(1, 4, 4)
plt.boxplot(vinhos['alcool'])
plt.title('Álcool')

plt.tight_layout()
plt.show()
```

```python
colunas_interesse = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']

plt.figure(figsize=(15, 10))

for i, coluna in enumerate(colunas_interesse):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='classe', y=coluna, data=vinhos)
    plt.title(f'Boxplot de {coluna} por Classe')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.title('Histograma da Distribuição de Fenois Totais')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais (Classe 3)')
plt.ylabel('Frequência')
plt.title('Histograma de Fenois Totais para a Classe 3')
plt.show()
```
