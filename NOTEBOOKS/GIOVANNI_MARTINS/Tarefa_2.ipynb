```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

> # **3. BOXPLOTS SEGREGADOS POR CLASSE**

```python
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de Magnesio por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Magnesio')
plt.show()
```

```python
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenois Totais por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Fenois Totais')
plt.show()
```

```python
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ácido Malico por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Ácido Malico')
plt.show()
```

```python
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Álcool')
plt.show()
```



> # **4. HISTOGRAMAS**



```python
plt.figure(figsize=(10, 6))
plt.hist(vinhos['fenois_totais'], bins=10, edgecolor='black')
plt.title('Distribuição dos dados de Fenois Totais')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python
vinhosClasse3 = vinhos[vinhos['classe'] == '3']
plt.figure(figsize=(10, 6))
plt.hist(vinhosClasse3['fenois_totais'], bins=10, edgecolor='black')
plt.title('Distribuição dos dados de Fenois Totais filtrados pela Classe 3')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```
