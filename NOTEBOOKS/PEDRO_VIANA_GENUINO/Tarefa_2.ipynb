```python
# importar bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

# Análise Exploratória de Variáveis

Contagem das classes presente na coluna:

```python
contagem_classes = vinhos['classe'].value_counts()
print("Contagem das classes:")
print(contagem_classes)
```

Estatísticas das colunas 'alcool', 'acido_malico', 'magnesio' e 'matiz':

```python
estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print("\nEstatísticas descritivas:")
print(estatisticas)
```


# Criação de Boxplots

```python
# Tamanho dos gráficos:

plt.figure(figsize=(24, 20))

# Boxplot 'magnesio'
plt.subplot(2, 2, 1)  # Criar subplots (Posição 1)
sns.boxplot(data=vinhos, y='magnesio')
plt.title('Boxplot de Magnesio')

# Boxplot 'prolina'
plt.subplot(2, 2, 2)  # Posição 2
sns.boxplot(data=vinhos, y='prolina')
plt.title('Boxplot de Prolina')

# Boxplot 'cinzas'
plt.subplot(2, 2, 3)  # Posição 3
sns.boxplot(data=vinhos, y='cinzas')
plt.title('Boxplot de Cinzas')

# Boxplot 'alcool'
plt.subplot(2, 2, 4)  # Posição 4
sns.boxplot(data=vinhos, y='alcool')
plt.title('Boxplot de Alcool')

# Ajustar o layout e exibir os gráficos:
plt.tight_layout()
plt.show()
```

# Criar Boxplots por classe

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 12))  # Cria uma grade de 2x2

# Boxplot 'magnesio' segregado por 'classe'
vinhos.boxplot(column='magnesio', by='classe', ax=axes[0, 0])
axes[0, 0].set_title('Boxplot de Magnesio por Classe')
axes[0, 0].set_xlabel('Classe')
axes[0, 0].set_ylabel('Magnesio')

# Boxplot 'fenois_totais' segregado por 'classe'
vinhos.boxplot(column='fenois_totais', by='classe', ax=axes[0, 1])
axes[0, 1].set_title('Boxplot de Fenois Totais por Classe')
axes[0, 1].set_xlabel('Classe')
axes[0, 1].set_ylabel('Fenois Totais')

# Boxplot 'acido_malico' segregado por 'classe'
vinhos.boxplot(column='acido_malico', by='classe', ax=axes[1, 0])
axes[1, 0].set_title('Boxplot de Ácido Málico por Classe')
axes[1, 0].set_xlabel('Classe')
axes[1, 0].set_ylabel('Ácido Málico')

# Boxplot 'alcool' segregado por 'classe'
vinhos.boxplot(column='alcool', by='classe', ax=axes[1, 1])
axes[1, 1].set_title('Boxplot de Álcool por Classe')
axes[1, 1].set_xlabel('Classe')
axes[1, 1].set_ylabel('Álcool')

# Ajustar o layout e exibir os gráficos:
plt.tight_layout()
plt.show()
```

# Criar Histograma

Visualização Fenois Totais:

```python
plt.figure(figsize=(10, 6))

# Histograma 'fenois_totais' com 10 bins
plt.hist(vinhos['fenois_totais'], bins=10, color='blue', alpha=0.7)

# Adicionar título e rótulos
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')

# Mostrar o gráfico adicionando linhas de grade
plt.grid(axis='y')
plt.show()
```

Visualização Fenois Totais segregado por classe:

```python

# Filtrar os dados para a classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

# Definir o tamanho da figura
plt.figure(figsize=(10, 6))

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(vinhos_classe_3['fenois_totais'], bins=10, color='blue', alpha=0.7)

# Criar o histograma da coluna 'fenois_totais' da classe 3 com 10 bins
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, color='blue', alpha=0.7)

# Adicionar título e rótulos
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.grid(axis='y')  # Adiciona linhas de grade para facilitar a leitura
plt.show()
```
