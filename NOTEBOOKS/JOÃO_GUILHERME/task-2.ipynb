```python

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

print(vinhos['alcool'].describe())
sns.histplot(data=vinhos, x='alcool', bins=20, kde=True)
plt.title('Distribuição do Teor Alcoólico')
plt.xlabel('Teor Alcoólico')
plt.ylabel('Frequência')
plt.show()

sns.boxplot(x='classe', y='alcool', data=vinhos)
plt.title('Comparação do Teor Alcoólico por Classe')
plt.xlabel('Classe')
plt.ylabel('Teor Alcoólico')
plt.show()

correlation = vinhos.corr()['alcool']
print(correlation)


print(vinhos['acido_malico'].describe())

sns.histplot(data=vinhos, x='acido_malico', bins=20, kde=True)
plt.title('Distribuição do Ácido Málico')
plt.xlabel('Ácido Málico')
plt.ylabel('Frequência')
plt.show()

sns.boxplot(x='classe', y='acido_malico', data=vinhos)
plt.title('Comparação do Ácido Málico por Classe')
plt.xlabel('Classe')
plt.ylabel('Ácido Málico')
plt.show()

correlation = vinhos.corr()['acido_malico']
print(correlation)


print(vinhos['magnesio'].describe())


sns.histplot(data=vinhos, x='magnesio', bins=20, kde=True)
plt.title('Distribuição do Magnésio')
plt.xlabel('Magnésio')
plt.ylabel('Frequência')
plt.show()


sns.boxplot(x='classe', y='magnesio', data=vinhos)
plt.title('Comparação do Magnésio por Classe')
plt.xlabel('Classe')
plt.ylabel('Magnésio')
plt.show()

correlation = vinhos.corr()['magnesio']
print(correlation)


print(vinhos['matiz'].describe())

sns.histplot(data=vinhos, x='matiz', bins=20, kde=True)
plt.title('Distribuição da Matiz')
plt.xlabel('Matiz')
plt.ylabel('Frequência')
plt.show()

sns.boxplot(x='classe', y='matiz', data=vinhos)
plt.title('Comparação da Matiz por Classe')
plt.xlabel('Classe')
plt.ylabel('Matiz')
plt.show()

correlation = vinhos.corr()['matiz']
print(correlation)

# Criação de boxplots

sns.boxplot(x='magnesio', data=vinhos)
plt.title('Boxplot da Variável Magnésio')
plt.xlabel('Magnésio')
plt.show()

sns.boxplot(x='prolina', data=vinhos)
plt.title('Boxplot da Variável Prolina')
plt.xlabel('Prolina')
plt.show()

sns.boxplot(x='cinzas', data=vinhos)
plt.title('Boxplot da Variável Cinzas')
plt.xlabel('Cinzas')
plt.show()

sns.boxplot(x='álcool', data=vinhos)
plt.title('Boxplot da Variável Álcool')
plt.xlabel('Álcool')
plt.show()

# Boxplot pelo objeto coluna
vinhos.boxplot(column='magnesio', by='classe')
vinhos.boxplot(column='fenois_totais', by='classe')
vinhos.boxplot(column='acido_malico', by='classe')
vinhos.boxplot(column='alcool', by='classe')

# Histograma para a tablea fenois_totais
sns.histplot(data=vinhos, x='fenois_totais', bins=10, kde=True)
plt.title('Distribuição dos Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()


#ultima atividade
df_classe_3 = vinhos[vinhos['classe'] == 3]
sns.histplot(data=df_classe_3, x='fenois_totais', bins=10, kde=True)
plt.title('Distribuição dos Fenóis Totais para a Classe 3')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()


```
