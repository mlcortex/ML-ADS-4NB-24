```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em portugus BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# 1. Anlise Exploratria de Variveis

# Contagem das diferentes classes presentes no dataset
contagem_classes = vinhos['classe'].value_counts()

# Estatsticas descritivas para as colunas solicitadas
estatisticas_alcool = vinhos['alcool'].describe()
estatisticas_acido_malico = vinhos['acido_malico'].describe()
estatisticas_magnesio = vinhos['magnesio'].describe()
estatisticas_matiz = vinhos['matiz'].describe()

# 2. Criao de Boxplots

# Configurar estilo dos grficos
sns.set(style="whitegrid")

# Boxplots das colunas solicitadas
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.boxplot(data=vinhos, y='magnesio')
plt.title('Boxplot - Magnsio')

plt.subplot(2, 2, 2)
sns.boxplot(data=vinhos, y='prolina')
plt.title('Boxplot - Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(data=vinhos, y='cinzas')
plt.title('Boxplot - Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(data=vinhos, y='alcool')
plt.title('Boxplot - lcool')

plt.tight_layout()
plt.show()

# 3. Boxplots Segregados por Classe

# Boxplots segregados por classe
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot - Magnsio por Classe')
plt.suptitle('')

plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot - Fenis Totais por Classe')
plt.suptitle('')

plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot - cido Mlico por Classe')
plt.suptitle('')

plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot - lcool por Classe')
plt.suptitle('')

plt.tight_layout()
plt.show()

# 4. Criao de Histogramas

# Histograma para fenois_totais com 10 bins
plt.figure(figsize=(12, 6))
plt.hist(vinhos['fenois_totais'], bins=10, color='blue', alpha=0.7)
plt.title('Histograma - Fenis Totais')
plt.xlabel('Fenis Totais')
plt.ylabel('Frequncia')
plt.show()

# Histograma para fenois_totais filtrando apenas a classe 3, com 10 bins
plt.figure(figsize=(12, 6))
plt.hist(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10, color='green', alpha=0.7)
plt.title('Histograma - Fenis Totais (Classe 3)')
plt.xlabel('Fenis Totais')
plt.ylabel('Frequncia')
plt.show()

# Retornar as estatsticas descritivas e a contagem das classes
contagem_classes, estatisticas_alcool, estatisticas_acido_malico, estatisticas_magnesio, estatisticas_matiz
```
